```python
#4.What percentage of smartphones have fast charging available?
total_fast_charging = no_outliers_df['fast_charging'].sum()
total_fast_charging
total_phones = len('no_outliers_df')
total_phones
(total_fast_charging,total_phones)
percent_of_fast_charging_phones = (total_fast_charging / total_phones).round(3) * 100
percent_of_fast_charging_phones

```

```python
#5.Analyzed the impact of processor brand on performance and rating.

processor_brand_stats = no_outliers_df.groupby('processor_brand')['rating'].mean().reset_index().sort_values('rating',ascending=False).round(3)
processor_brand_stats

```

```python
#Analyze the impact of processor brand on performance and rating.
sns.barplot(data =processor_brand_stats , y ='processor_brand', x = 'rating')
plt.title('Precessor brand performances')
plt.show
```

```python
#6.Analyze the performance differences between smartphones with different operating systems.
# Analyze the performance differences by operating system
os_performance_stats = no_outliers_df.groupby('os')[['price','rating']].agg(['mean','sum']).reset_index()
# Renaming columns for clarity
os_performance_stats.columns = ['operating_system', 'average_price', 'total_price', 'average_rating', 'total_rating']
# Sorting values
os_performance_stats = os_performance_stats.sort_values(['average_price', 'total_price', 'average_rating', 'total_rating'], ascending=False)
os_performance_stats
```

```python
#Visualized average price for each operating system.
sns.barplot(data =os_performance_stats, x = 'operating_system' , y = 'average_price')
plt.title('Average Price by operating system')
plt.show
```

```python
sns.barplot(data =os_performance_stats, x = 'operating_system' , y = 'total_price')
plt.title('Total Price by Operating system')
plt.show
```

```python
sns.barplot(data =os_performance_stats, x = 'operating_system' , y = 'average_rating')
plt.title('Operating system by average rating')
plt.show
```

```python
sns.barplot(data =os_performance_stats, x = 'operating_system' , y = 'total_rating')
plt.title('Total rating by Operating system')
plt.show
```

```python
#7.How does processor speeds and the number of cores vary across different processor brands?
processor_stats = no_outliers_df.groupby('processor_brand')[['processor_speed','num_cores']].agg(['mean', 'count']).reset_index()
processor_stats.columns = ['processor_brand','average_speed','num_of_precessor','average_num_cores','num_of_cores']
processor_stats.columns 
performance_dist = processor_stats.columns .sort_values(['processor_brand','average_speed','num_of_precessor','average_num_cores','num_of_cores'], ascending=False)
performance_dist
(processor_stats)
```

```python
#Visualized average price for each operating system.
sns.regplot(data =processor_stats, y = 'average_speed' , x = 'num_of_precessor')
plt.title('Relationship by avg speed and number of precessor')
plt.show
```

```python
#8.What was the average price of smartphones model with 5G? 
# Filter the DataFrame for 5G smartphones
smartphones_with_5g = no_outliers_df[no_outliers_df['has_5g'] == True]

# Calculate the average price of 5G smartphones
average_price_5g = smartphones_with_5g['price'].mean().round(2)

average_price_5g
# Group by model for 5G smartphones and calculate the mean price
average_price_5g_per_model = smartphones_with_5g.groupby('model')['price'].mean().reset_index().sort_values('price',ascending=False)
average_price_5g_per_model.head

#Find top ten model price
top_ten_model_price = average_price_5g_per_model.nlargest(10,'price')
top_ten_model_price

```

```python
#Visualized average price for smartphone model with 5G.
sns.barplot(data=top_ten_model_price, y = 'model' , x = 'price')
plt.title('Average price by model with 5g')
plt.show
```

```python
#9.What was the average price of smartphones model without 5G? 
# Filter the DataFrame for 5G smartphones
smartphones_without_5g = no_outliers_df[no_outliers_df['has_5g'] == False]

# Calculate the average price of model without 5G smartphones
average_price_without_5g = smartphones_with_5g['price'].mean().round(2)

average_price_without_5g
# Group by model for 5G smartphones and calculate the mean price
average_price_without_5g = smartphones_with_5g.groupby('model')['price'].mean().reset_index().sort_values('price',ascending=False)

average_price_without_5g.head(30)

#Find top ten model price
top_ten_model_price = average_price_5g_per_model.nlargest(10,'price')
top_ten_model_price
```

```python
#Visualized average price for smartphone model without 5G.
sns.barplot(data=top_ten_model_price, y = 'model' , x = 'price')
plt.title('Average price by model without 5g')
plt.show
```
