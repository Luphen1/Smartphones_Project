```python
#Checked and removed outliers using IQR(interquatile Range)


Q3 = 35491.50
Q1 = 12999.000
Q3,Q1
```

```python
IQR = Q3 - Q1
IQR = 35491.50 - 12999.000
IQR 
```

```python
lower_limit = Q1 - IQR * 1.5
lower_limit = 12999.000 - 22492.5 * 1.5
lower_limit
```

```python
upper_limit = Q3 + IQR * 1.5
upper_limit = 35491.50 + 22492.5 * 1.5
upper_limit
```

```python
no_outliers_df = df[(df.price > lower_limit) & (df.price < upper_limit)]
no_outliers_df
```

```python
#After removing outliers
plt.figure(figsize=(10,6))
sns.boxplot(data = no_outliers_df , y = 'price')
plt.title('Cleaned Dataframe')
plt.show
```

```python
#Ready for Exploratory Data Analysis.
#During the analysis I would be working with 'no_outliers_df'.

#1.What are the smartphone prices across different top ten brands?
dist_price =  no_outliers_df.groupby('brand_name')['price'].sum().reset_index().nlargest(10,'price')
dist_price
```

```python
# Visualized output.
plt.figure(figsize=(10,6))
sns.barplot(data = dist_price, y ='brand_name', x ='price')
plt.title('Distribution price by brand')
plt.show
```

```python
#2.Are there any correlation between the price of a smartphone and its battery capacity,RAM capacity and internal memory,and the price and rating.

corr_relationship = no_outliers_df[['fast_charging','battery_capacity','ram_capacity','internal_memory','price','rating']].corr()
corr_relationship
```

```python
#Visualized output.
plt.figure(figsize=(10,6))
sns.heatmap(corr_relationship, annot=True, cmap='Spectral', fmt=".2f")
plt.title('Correlation Relationship')
plt.show()
```

```python
#3.What are the price distribution across the dataset?
#Visualized price distribution across the dataframe
plt.figure(figsize=(10,6))
sns.histplot(data = no_outliers_df , x ='price',kde=True,color='green')
plt.title('Price distribution')
plt.show
```

```python
#Statistic information within the dataframe
correlation_relationship = no_outliers_df.describe().T
correlation_relationship
```

```python
#4.What are the percentage of smartphones with fast charging available?

num_of_fast_charging = no_outliers_df['fast_charging_available'].shape[0]
num_of_fast_charging
percent_of_fast_charging_available = (no_outliers_df.groupby('brand_name')['fast_charging_available'].sum() / num_of_fast_charging *100).reset_index().sort_values('fast_charging_available',ascending=False)
percent_of_fast_charging_available

```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data =percent_of_fast_charging_available , y ='brand_name', x = 'fast_charging_available')
plt.title('Percentage of smart charging available')
plt.ylabel('brand-name')
plt.xlabel('fast_charging')
plt.show
```

```python
#5.How does the processor brand impact performance and rating?

processor_brand_stats = no_outliers_df.groupby('processor_brand')['rating'].mean().reset_index().sort_values('rating',ascending=False).round(3)
processor_brand_stats

```

```python
#Analyze the impact of processor brand on performance and rating.
plt.figure(figsize=(10,6))
sns.barplot(data =processor_brand_stats , y ='processor_brand', x = 'rating')
plt.title('Precessor brand performances')
plt.show
```

```python
#6.What are the performance differences among different operating systems?
os_performance_stats = no_outliers_df.groupby('os')[['price','rating']].agg(['mean','sum']).reset_index()
# Renaming columns for clarity
os_performance_stats.columns = ['operating_system', 'average_price', 'total_price', 'average_rating', 'total_rating']
# Sorting values
os_performance_stats = os_performance_stats.sort_values(['average_price', 'total_price', 'average_rating', 'total_rating'], ascending=False)
os_performance_stats
```

```python
#Visualized average price for each operating system.
plt.figure(figsize=(10,6))
sns.barplot(data =os_performance_stats, x = 'operating_system' , y = 'average_price')
plt.title('Average Price by operating system')
plt.show
```

```python
#Visualized total price for each operating system.
plt.figure(figsize=(10,6))
sns.barplot(data =os_performance_stats, x = 'operating_system' , y = 'total_price')
plt.title('Total Price by Operating system')
plt.show
```

```python
#Visualized average rating for each operating system.
plt.figure(figsize=(10,6))
sns.barplot(data =os_performance_stats, x = 'operating_system' , y = 'average_rating')
plt.title('Operating system by average rating')
plt.show
```

```python
#Visualized total rating for each operating system.
plt.figure(figsize=(10,6))
sns.barplot(data =os_performance_stats, x = 'operating_system' , y = 'total_rating')
plt.title('Total rating by Operating system')
plt.show
```

```python
#7.How does processor speeds and the number of cores vary across different processor brands?
processor_stats = no_outliers_df.groupby('processor_brand')[['processor_speed','num_cores']].mean().reset_index()
processor_stats 
performance_diff = processor_stats.sort_values(['processor_speed','num_cores'], ascending=False)
performance_diff
```

```python
#Visualized the relationships between num_cores and processor speed
plt.figure(figsize=(10,6))
sns.regplot(data =processor_stats, y = 'num_cores' , x = 'processor_speed')
plt.title('Relationship between num_cores and processor_speed ')
plt.show
```

```python
#Visualized output
plt.figure(figsize=(10,6))
sns.barplot(data =processor_stats, y = 'processor_brand' , x = 'processor_speed')
plt.title('processor brand by speed')
plt.show
```

```python
#Visualized output
plt.figure(figsize=(10,6))
sns.barplot(data =processor_stats, y = 'processor_brand' , x = 'num_cores')
plt.title('Processor brand by number of cores')
plt.show
```

```python
#8.What were the top five average price of smartphones model with 5G? 
# Filter the DataFrame for 5G smartphones
smartphones_with_5g = no_outliers_df[no_outliers_df['has_5g'] == True]

# Calculate the average price of 5G smartphones
average_price_5g = smartphones_with_5g['price'].mean().round(2)

average_price_5g
# Group by model for 5G smartphones and calculate the mean price
average_price_5g_per_model = smartphones_with_5g.groupby('model')['price'].mean().reset_index().sort_values('price',ascending=False)
average_price_5g_per_model.head

#Find top ten model price
top_ten_model_price = average_price_5g_per_model.nlargest(10,'price')
top_ten_model_price

```

```python
#Visualized average price for smartphone model with 5G.
plt.figure(figsize=(10,6))
sns.barplot(data=top_ten_model_price, y = 'model' , x = 'price')
plt.title('Average price by model with 5g')
plt.show
```
